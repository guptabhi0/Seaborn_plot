```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fmri = sns.load_dataset("fmri")
fmri.head()
```

```python
sns.lineplot(x='timepoint',y = 'signal', data = fmri)
```

```python
sns.lineplot(x='timepoint',y = 'signal', data = fmri, hue = 'event')
```

```python
plt.show()
```

```python
sns.lineplot(x='timepoint',y = 'signal', data = fmri, hue = 'event',style = 'event', markers = True)
```

```python
import pandas as pd
```

```python
sns.set(style= 'whitegrid')
```

```python
car = pd.read_csv('car data.csv')
```

```python
car.head()
```

```python
sns.barplot(y = "Present_Price", x = "Selling_Price",  hue= "Transmission", data = car,color = 'red')
plt.show()
```

```python
sns.scatterplot(y = "Present_Price", x = "Selling_Price", hue = "Transmission",data= car)
```

```python
sns.distplot(car["Present_Price"])
```

```python
sns.distplot(car["Selling_Price"], hist = False, color = 'green')
```

```python
sns.distplot(car["Selling_Price"], kde = False, color = 'green', bins= 100)
```

```python
sns.distplot(car["Selling_Price"], kde = False, color = 'green', bins= 5)
```

```python
sns.distplot(car["Selling_Price"], kde = False, color = 'green', bins= 20, vertical = True)
```

```python
sns.jointplot(y = "Present_Price", x = "Selling_Price",data= car)
```

```python
sns.jointplot(y = "Present_Price", x = "Selling_Price",data= car,kind = 'reg')
```

```python
sns.boxplot(y = "Present_Price", x = "Selling_Price",data= car)
```
